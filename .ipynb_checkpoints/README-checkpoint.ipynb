StoryWeaver: Un Generador de Cuentos Infantiles con IA
Resumen del Proyecto
StoryWeaver es un proyecto interactivo desarrollado en un Jupyter Notebook que fusiona la creatividad de la inteligencia artificial con la narrativa infantil. Utiliza el modelo de lenguaje Gemini 1.5 Flash para generar cuentos personalizados y didácticos, así como descripciones detalladas para ilustrar la primera escena de cada historia. Estas descripciones están optimizadas para ser usadas en herramientas de IA de generación de imágenes como Nightcafe.

Este proyecto es una prueba de concepto (POC) funcional que demuestra cómo la IA puede ser utilizada como una herramienta para la creación de contenido educativo y artístico.

Características Principales
Creación de Cuentos Personalizados: Genera una historia única a partir de un tema y un personaje proporcionados por el usuario.

Generación de Prompts de Imagen: Crea descripciones optimizadas, concisas y listas para ser usadas en modelos de IA de texto a imagen.

Diseño Didáctico: Las historias son simples, positivas y están diseñadas para niños pequeños.

Flujo de Trabajo Simplificado: Todo el proceso de generación de texto y prompts se lleva a cabo en un solo script de Python.

Requisitos
Para ejecutar el proyecto, necesitas tener instalado lo siguiente:

Python 3.x

Jupyter Notebook o JupyterLab

Guía de Instalación y Uso
Sigue estos pasos para poner en marcha el proyecto:

Clonar el Repositorio (Opcional): Si tienes el proyecto en un repositorio, clona el repositorio en tu máquina local.

Instalar las Dependencias: Abre tu terminal o la línea de comandos y ejecuta el siguiente comando para instalar las bibliotecas necesarias:

pip install google-generativeai ipython

Si usas un entorno de Jupyter, puedes ejecutar este comando directamente en una celda:

!python -m pip install google-generativeai ipython

Obtener tu Clave de API de Gemini:

Ve a Google AI Studio (https://aistudio.google.com/).

Crea un nuevo proyecto y genera tu clave de API.

Copia la clave.

Configurar la Clave de API: Para que el script pueda acceder a la API, debes configurar la clave como una variable de entorno.

En macOS/Linux: Abre tu terminal y ejecuta:

export GEMINI_API_KEY="TU_CLAVE_AQUI"

En Windows (Símbolo del sistema):

set GEMINI_API_KEY="TU_CLAVE_AQUI"

En Windows (PowerShell):

$env:GEMINI_API_KEY="TU_CLAVE_AQUI"

Ejecutar el Notebook: Abre Jupyter Notebook y navega al archivo StoryWeaver_limpio.py.

Interactuar con el Cuento: Ejecuta las celdas una por una. El script te pedirá que ingreses el tema y el personaje de tu historia.

Tecnologías Utilizadas
Python: Lenguaje de programación.

Google Gemini 1.5 Flash: Modelo de lenguaje generativo.

Nightcafe: Herramienta de generación de imágenes por IA (uso manual).

Jupyter Notebook: Entorno de desarrollo.
